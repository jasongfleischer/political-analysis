I read the following article https://www.vox.com/policy-and-politics/2020/2/7/21123518/trump-2020-election-democratic-party-james-carville and it contained a quote from James Carville that I didn't quite believe:

    “Eighteen percent of the population controls 52 Senate seats”
    
It couldn't be that bad, could it?  That 18% controls more than half the senatorial power?

```python
import pandas as pd
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
spops = pd.read_csv('https://www2.census.gov/programs-surveys/popest/datasets/2010-2019/national/totals/nst-est2019-alldata.csv')
```

```python
spops.head(15)
```

```python
# note that we need to get rid of several items in this table, 
# since they are regional summaries or non-senatorial places like PR and DC
# also let's use only NAME, CENSUS2010POP and POPESTIMATE2019
# sort it by smallest to biggest states so we see the worst case scenario of 
# how few people are rep'ed by 52 senators
statecdf = spops.set_index('NAME').drop(['United States','Northeast Region', 'Midwest Region', 
                                      'South Region', 'West Region',
                                      'District of Columbia', 'Puerto Rico']
                                    )[['CENSUS2010POP','POPESTIMATE2019']].sort_values('POPESTIMATE2019')
statecdf['CumPop%2019'] = statecdf['POPESTIMATE2019'].cumsum() / statecdf['POPESTIMATE2019'].sum() * 100 
statecdf['CumPop%2010'] = statecdf['CENSUS2010POP'].cumsum() / statecdf['CENSUS2010POP'].sum() * 100 
# note that these sums are % of the population that HAVE SENATORS (does not include DC/PR/Territories) !
```

```python
statecdf.iloc[26] # the 51st and 52nd senators come from state #26
```

```python
statecdf.index[:27] # the smallest states by 2019 population
```

Wankhammer douchenozzle! Thats ~19% no matter if you use 2010 or 2019 data. It's actually roughly true!

Now Carville was probably talking about the Republican seats, not the smallest 26 states.  This list has a few recognizably Democratic or mixed states in it.  I make it (by eye, not looked up) about 1/3 of these senators are D, maybe a bit less.  So assumedly, either Carville is exaggerating or maybe (incorrectly in my view) normalizing this by the population of the entire US (incl. PR/DC) to get to 18%.  


I am very curious about the historical trend of this data... was it always so imbalanced or is this a recent phenomena???

However, sadly the census website doesn't have the data in this state based format before 2000.  

I did find various online sources for state congressional apportionment or population, but only in text tables, not csv or some such.  So I copy/pasted the values from wikipedia's 1960 census <https://en.wikipedia.org/wiki/1960_United_States_Census> into an excel file and calculated the bottom 26 state's population percentage compared to the whole:  18.7%

In 1930 (only 48 states, so I took bottom 25 states populations) it was 17.6%

Another way to get at this is the apportionment of congressional seats... it's got problems (like granularity of the estimate, the old 3/5 for African Americans, not counting women before a certain point), but at least I can download the numbers and the estimate of state population from it isn't completely terrible.


```python
scong = pd.read_csv('https://www.brookings.edu/wp-content/uploads/2017/01/vitalstats_ch1_tbl1.csv').drop('Region',axis=1).set_index('State')
```

```python
scong.head(15)
```

```python
dlist = [ el for el in scong.index if ('All' in el)] # get rid of regional summaries, leaving only states
scong = scong.drop( dlist )
print(scong.groupby('Year')['Seats'].sum()) # number of seats per congress, blip in 1950 for admission of AL and HI
```

```python
vals = []

for ayear in scong.Year.unique():
    thiscong = scong[ scong.Year == ayear ].sort_values('Seats')
    thiscong['Cum%Pop for 1 state more than half of the senate'] = thiscong.Seats.cumsum() * 1./thiscong.Seats.sum() * 100
    vals.append( thiscong.iloc[ int(round(thiscong.shape[0]/2))+1, : ] ) 

pd.DataFrame(vals ).set_index('Year').drop('Seats',axis=1)
```

It appears the congressional seat proxy work pretty well, as the 2000/2010 numbers agree with the census numbers.

So this does appear to be a long term feature of our system, not necessarily one that is getting worse as we become more urban.  It looks like we dropped a few perent from the turn of the 20th century, but I don't see it getting noticably worse since 1950.
